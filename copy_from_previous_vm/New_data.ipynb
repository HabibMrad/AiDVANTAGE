```python
import os
import pandas as pd

def DataInit():#Инициализация данных
    path = os.curdir + '/Final_data/'
    DataSetName = os.listdir(path) #Получение сведений о содержимом в папке
    temp = []
    for el in DataSetName: #Цикл поиска эллементов без .csv
        if (el.find('.csv') == -1): #Если в элементе нет .csv то функция .find() выдает -1
            temp.append(el)

    for el in temp:#цикл удаление элементов без .csv
        DataSetName.remove(el)
        
    files = []
    
#     for el in DataSetName:
#         #frames.append(pd.read_csv(path + el))
#     #return pd.concat(frames, ignore_index=True)

    for el in DataSetName:
        files.append(el.replace('.csv',''))
        #frames.append(pd.read_csv(path + el))
    #return pd.concat(frames, ignore_index=True)
    
    files.sort(key=int) 
    
    frames = []
    
    for el in files:
        frames.append(pd.read_csv(path + el + '.csv').transpose().iloc[1:])
    return pd.concat(frames, sort=False, ignore_index=True)

LastCol = DataInit()
LastCol.rename(columns={0:'Диагноз'}, inplace=True)
LastCol = LastCol[LastCol.Диагноз != 'rinat1nuriev@gmail.com']
LastCol = LastCol[LastCol.Диагноз != 'alex.boroda@lenta.ru']
```
