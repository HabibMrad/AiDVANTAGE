```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import datetime
```

```python
df = pd.read_csv('For delete/ready_to_work.csv')

df_x = df['4.Болели ли Ваши кровные родственники злокачественными опухолевыми заболеваниями? ']
df_y = df['6.Вы курите?']
df_z = df['Диагноз']
```

```python
fig = go.Figure(data=go.Heatmap(x=df_x.values, y=df_y.values, z=df_z.values))
# fig.show()

fig.update_layout(
    autosize=False,
    width=500,
    height=500,
    xaxis=go.layout.XAxis(
        title_text = df_x.name,
        ticktext=["Нет", "Да", "Не знаю"],
        tickvals=[0, 1, 2],
        tickmode="array",
        titlefont=dict(size=10),
    ),
    yaxis=go.layout.YAxis(
        title_text=df_y.name,
        ticktext=["Нет, никогда не курил", "Да, больше 10 лет", "Да, меньше 10 лет"],
        tickvals=[0, 1, 2],
        tickmode="array",
        titlefont=dict(size=10),
    )
)

```

```python
fig = go.Figure()

fig.add_trace(go.Bar(
    x=df_z,
    y=df_x
))

fig.update_layout(
    autosize=False,
    width=500,
    height=500,
    xaxis=go.layout.XAxis(
        title_text="Параметр",
        ticktext=["Низкий риск", "Средний риск", "Высокий риск"],
        tickvals=[0, 1, 2],
        tickmode="array",
        titlefont=dict(size=30),
    ),
    yaxis=go.layout.YAxis(
        title_text="Диагноз",
#         ticktext=["Низкий риск", "Средний риск", "Высокий риск"],
#         tickvals=[1, 2, 3],
        tickmode="array",
        titlefont=dict(size=30),
    )
)

fig.update_yaxes(automargin=True)

fig.show()
```
